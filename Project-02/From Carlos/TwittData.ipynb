```python
# do a loop and check objects 
#results_test = api.GetSearch("bitcoin", since=2018-12-15, count=500)
results_test = api.GetSearch("bitcoin", count=500)
for tweet in results_test:
    print(tweet.id, tweet.text)
    
```

```python
#print(tweet.id, tweet.text, tweet.created_at, tweet.user.screen_name, tweet.source)
#print(tweet.favorite_count, tweet.text, tweet.retweet_count)
topics_dict =   {"id":[], 
                "tweet":[],
                "screenName":[], 
                "url":[], 
                "created":[], 
                "retweet":[], 
                "favorite_count":[]}
```

```python
for tweet in results_test:
    topics_dict["id"].append(tweet.id)
    topics_dict["tweet"].append(tweet.text)
    topics_dict["screenName"].append(tweet.user.screen_name)
    topics_dict["url"].append(tweet.source)
    topics_dict["created"].append(tweet.created_at)
    topics_dict["retweet"].append(tweet.retweet_count)
    topics_dict["favorite_count"].append(tweet.favorite_count)
    
```

```python
topics_data = pd.DataFrame(topics_dict)
topics_data

#topics_data.sort_values("favorite_count", ascending=False)
topics_data.sort_values(["retweet", "favorite_count"], ascending=[False, False])
```

```python
# Dependencies
# ----------------------------------
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
```

```python
from sqlalchemy import Column, Integer, String, Float, UnicodeText, Date, BigInteger
```

```python
# Create Bitcoin_Twitter Classe:'ScreenName', 'Tweet', 'retweet', 'favorite_count', 'Created', source
# ----------------------------------


class Bitcoin_Twitter(Base):
    __tablename__ = 'bitcoin_twitter'
    id = Column(Integer, primary_key=True)
    screenName = Column(String(100))
    tweet = Column(UnicodeText)
    retweet = Column(Integer)
    favorite_count = Column(Integer)
    created = Column(Date)
    source = Column(String(20))
```

```python
# Add Records to the Appropriate DB
# ----------------------------------
# Use the SQL ALchemy methods to run simple "INSERT" statements using the class and object of Bitcoin_Twitter.

for i in range(len(new_topics_data_df['tweet'])):  
    one_bitcoin_tweet = Bitcoin_Twitter(screenName=new_topics_data_df.loc[i,'screenName'],
                                    tweet=new_topics_data_df.loc[i,'tweet'],
                                    source=new_topics_data_df.loc[i,'source'],
                                    retweet=new_topics_data_df.loc[i,'retweet'],
                                    favorite_count=new_topics_data_df.loc[i,'favorite_count'],
                                    created=new_topics_data_df.loc[i,'created'],)
    session.add(one_bitcoin_tweet)
    session.commit()
```
